```python
'''
Created on Dec 24, 2017

Cryptocurrency Price Alert
This is a simple code for scraping the Coinmarketcap website, getting latest crypto price and send a text to the user if a certain threshold is met

@author: pedram oskouie
'''
import requests
import smtplib
from bs4 import BeautifulSoup

# specify the url
page = requests.get("https://coinmarketcap.com/all/views/all/")

#A status code of 200 means successful page download
print (page.status_code)

symbol = []
market_cap = []
price = []
circulating_supply = []
volume = []


# parse the html using beautiful soup and store in variable `soup`
soup = BeautifulSoup(page.content, 'html.parser')

### - Find the div with id seven-day-forecast, and assign to seven_day
### - Inside seven_day, find each individual forecast item.
### - Extract and print the first (here is the third because of the website's condition) forecast item

### Bitcoin
bitcoin = soup.find(id="id-bitcoin")
bitcoin_fields = bitcoin.find_all(class_="no-wrap")
random_field = bitcoin_fields[2]
print(len(bitcoin_fields))

#We can now print out the HTML content of the page, formatted nicely, using the prettify method on the BeautifulSoup object
print(random_field.prettify())
print(bitcoin.prettify())

#Extracting all the fields

symbol.append (bitcoin.find(class_="col-symbol").get_text())
market_cap.append ( bitcoin.find(class_="market-cap").get_text())
price.append (bitcoin.find(class_="price").get_text())
circulating_supply.append (bitcoin.find(class_="circulating-supply").get_text())
volume.append (bitcoin.find(class_="volume").get_text())

print(symbol)
print(market_cap)
print(price)
print(circulating_supply)
print(volume)

### Ethereum
ether = soup.find(id="id-ethereum")
symbol.append (ether.find(class_="col-symbol").get_text())
market_cap.append(ether.find(class_="market-cap").get_text())
price.append(ether.find(class_="price").get_text())
circulating_supply.append(ether.find(class_="circulating-supply").get_text())
volume.append(ether.find(class_="volume").get_text())

### Bitcoin Cash
bit_cash = soup.find(id="id-bitcoin-cash")
symbol.append (bit_cash.find(class_="col-symbol").get_text())
market_cap.append(bit_cash.find(class_="market-cap").get_text())
price.append(bit_cash.find(class_="price").get_text())
circulating_supply.append(bit_cash.find(class_="circulating-supply").get_text())
volume.append(bit_cash.find(class_="volume").get_text())

### Ripple
ripple = soup.find(id="id-ripple")
symbol.append (ripple.find(class_="col-symbol").get_text())
market_cap.append(ripple.find(class_="market-cap").get_text())
price.append(ripple.find(class_="price").get_text())
circulating_supply.append(ripple.find(class_="circulating-supply").get_text())
volume.append(ripple.find(class_="volume").get_text())

### Litecoin

lite_coin = soup.find(id="id-litecoin")
symbol.append (lite_coin.find(class_="col-symbol").get_text())
market_cap.append(lite_coin.find(class_="market-cap").get_text())
price.append(lite_coin.find(class_="price").get_text())
circulating_supply.append(lite_coin.find(class_="circulating-supply").get_text())
volume.append(lite_coin.find(class_="volume").get_text())
### Stellar
stellar = soup.find(id="id-stellar")
symbol.append (stellar.find(class_="col-symbol").get_text())
market_cap.append(stellar.find(class_="market-cap").get_text())
price.append(stellar.find(class_="price").get_text())
circulating_supply.append(stellar.find(class_="circulating-supply").get_text())
volume.append(stellar.find(class_="volume").get_text())

### Qtum
qtum = soup.find(id="id-qtum")
symbol.append (qtum.find(class_="col-symbol").get_text())
market_cap.append(qtum.find(class_="market-cap").get_text())
price.append(qtum.find(class_="price").get_text())
circulating_supply.append(qtum.find(class_="circulating-supply").get_text())
volume.append(qtum.find(class_="volume").get_text())

print (len(symbol))
print (len(market_cap))
print(len(price))
print (len(circulating_supply))
print(len(volume))

### We can now combine the data into a Pandas DataFrame and analyze it. 
### A DataFrame is an object that can store tabular data, making data analysis easy. 
import pandas as pd
crypto_data = pd.DataFrame({
        "Symbol": symbol,
        "Price": price, 
        "Market_cap": market_cap,  
        "Circulating_supply": circulating_supply,
        "Volume": volume
    })
print(crypto_data)


###Saving the Panda into an Excel File

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('crypto_data.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
crypto_data.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()

# we import the Twilio client from the dependency we just installed
from twilio.rest import Client

# the following line needs your Twilio Account SID and Auth Token
client = Client("AC73f03cd701fed934d982d9c2xxxxxxxxx", "a087c6cb525d6fa51efb9799xxxxxxx")

# change the "from_" number to your Twilio number and the "to" number
# to the phone number you signed up for Twilio with, or upgrade your
# account to send SMS to any phone number

#Price Alert for Bitcoin
bitcoin_price = price[0]
bit_split = bitcoin_price.split("$")
bit_split = bit_split[1]
bitcoin_float = float(bit_split)

if (bitcoin_float<12000):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE DROP ALERT From Pedram Oskouie: The latest Bitcoin price is: " + price[0])

if (bitcoin_float>20000):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE SURGE ALERT From Pedram Oskouie: The latest Bitcoin price is: " + price[0])

#Price Alert for Ethereum
ether_price = price[1]
ether_split = ether_price.split("$")
ether_split = ether_split[1]
ether_float = float(ether_split)

if (ether_float<600):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE DROP ALERT From Pedram Oskouie: The latest Ethereum price is: " + price[1])

if (ether_float>1000):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE SURGE ALERT From Pedram Oskouie: The latest Ethereum price is: " + price[1])

#Price Alert for LiteCoin
lite_price = price[4]
lite_split = lite_price.split("$")
lite_split = lite_split[1]
lite_float = float(lite_split)

if (lite_float<300):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE DROP ALERT From Pedram Oskouie: The latest LiteCoin price is: " + price[4])

if (lite_float>1000):
    client.messages.create(to="+1213xxxxxxx", 
                       from_="+13237452292", 
                       body="PRICE SURGE ALERT From Pedram Oskouie: The latest LiteCoin price is: " + price[4])

```
